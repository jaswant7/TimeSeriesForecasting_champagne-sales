```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
series = pd.read_csv("./monthly_champagne_sales.csv",header=0,index_col=0,parse_dates=True, squeeze=True)
series.head()
```

```python
split_point = len(series) - 12
dataset, validation = series[0:split_point],series[split_point:]
print(len(dataset),len(validation))
```

## Data Analysis

```python
#summary statistics
dataset.describe()
```

```python
#line plot
dataset.plot()
plt.show()

#1.The seasonality appears to be growing over time
#2. The series is non-stationary
#3. Increasing Trend of sales over time.
```

```python
#seasonal line plots
groups = dataset['1964':'1970'].groupby(pd.Grouper(freq='A'))
years = pd.DataFrame()

for name,group in groups:
    years[name] = group.values
years.plot(subplots=True,legend=False)
plt.show()

#1. We can see rise in last months in all years
```
