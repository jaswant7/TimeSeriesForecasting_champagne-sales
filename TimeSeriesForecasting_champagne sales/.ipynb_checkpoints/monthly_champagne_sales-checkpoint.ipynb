```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
series = pd.read_csv("./monthly_champagne_sales.csv",header=0,index_col=0,parse_dates=True, squeeze=True)
series.head()
```

```python
split_point = len(series) - 12
dataset, validation = series[0:split_point],series[split_point:]
print(len(dataset),len(validation))
```

## Data Analysis

```python
#summary statistics
dataset.describe()
```

```python
#line plot
dataset.plot()
plt.show()

#1.The seasonality appears to be growing over time
#2. The series is non-stationary
#3. Increasing Trend of sales over time.
```

```python
#seasonal line plots
groups = dataset['1964':'1970'].groupby(pd.Grouper(freq='A'))
years = pd.DataFrame()

for name,group in groups:
    years[name] = group.values
years.plot(subplots=True,legend=False)
plt.show()

#1. We can see rise in last months in all years
```

```python
#Density plot
plt.figure(1)
plt.subplot(211)
dataset.hist()
plt.subplot(212)
dataset.plot(kind="kde")
plt.show()
#1. Not Gaussian distribution
#2. the shape has long tail
#3. power transformation can be applied
```

```python
groups = dataset['1964':'1970'].groupby(pd.Grouper(freq='A'))
years = pd.DataFrame()

for name,group in groups:
    years[name.year] = group.values
years.boxplot()
plt.show()
#1. increasing trend as medain is inc
#2. outliers are there
#3. last year look different
```

```python
#As data is not stationary, as seasonality is present
#we can diff the data from previous year's month  to make data stationary
X = dataset.values

diff = list()
for i in range(12,len(X)):
    diff.append(X[i]-X[i-12])

stationary = pd.Series(diff)

stationary.index = dataset.index[12:]

```

```python
#Now we will check for stationarity
from statsmodels.tsa.stattools import adfuller
result = adfuller(stationary)

print("ADF statistic",result[0])
print("p value",result[1])
print("critical values")
for key,value in result[4].items():
    print(key,value)
#we can select d=0 for ARIMA model
```

```python
#plot stationary series
stationary.plot()
plt.show()
```

```python
#ACF and PACF plots of differenced time series
from statsmodels.graphics.tsaplots import plot_acf
from statsmodels.graphics.tsaplots import plot_pacf

plt.figure()
plt.subplot(211)
plot_acf(stationary,lags=15,ax=plt.gca())
plt.subplot(212)
plot_pacf(stationary,lags=15,ax=plt.gca())
plt.show()
```

```python
#d=0,p=1,q=1 ARIMA(1,0,1)
from statsmodels.tsa.arima_model import ARIMA
from sklearn.metrics import mean_squared_error
from math import sqrt

def difference(history,interval):
    diff = list()
    for i in range(interval,len(history)):
        diff.append(history[i]-history[i-interval])
        
    return diff
def inv_diff(history,yhat,interval):
    return yhat + history[-interval]
X = dataset.values
split = int(len(X)*0.5)
train,test = X[0:split],X[split:]
history = [x for x in train]
predictions = []
for i in range(len(test)):
    diff = difference(history,12)
    model = ARIMA(diff,order=(1,1,1))
    model_fit = model.fit(trend='nc',disp=0)
    yhat = model_fit.forecast()[0]#standard error,CI are ignored
    yhat = inv_diff(history,yhat,12)
    predictions.append(yhat)
    obs = test[i]
    history.append(obs)
    print("predicted",yhat,"expected",obs)
rmse = sqrt(mean_squared_error(test,predictions))
print("rmse",rmse)
```

```python
plt.plot(test)
plt.plot(predictions, color='red')
plt.show()
```
